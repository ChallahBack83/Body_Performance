```python
import os
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("body_performance_data.csv")
df.head()
```

```python
df.info()
```

```python
df.reset_index(inplace=True)
df = df.rename(columns = {"index":"ID"})
df.head()
```

```python
Physical_metrics_df = pd.DataFrame(df[["ID", "age", "gender", "height_cm", "weight_kg", "body fat_%", "diastolic", "systolic"]])
Physical_metrics_df.head()
```

```python
Physical_metrics_df = Physical_metrics_df.rename(columns = {"body fat_%" :"body_fat_%"})
Physical_metrics_df.head(10)
```

```python
Activity_metrics_df = pd.DataFrame(df[["ID", "gripForce", "sit and bend forward_cm", "sit-ups counts", "broad jump_cm", "class"]])
Activity_metrics_df.head()
```

```python
Activity_metrics_df = Activity_metrics_df.rename(columns = {"sit and bend forward_cm" :"sit_and_bend_forward_cm", "sit-ups counts" :"sit_ups_counts", "broad jump_cm" :"broad_jump_cm"})
Activity_metrics_df.head(10)
```
